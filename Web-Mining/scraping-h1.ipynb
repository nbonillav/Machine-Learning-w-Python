## Parte 1
- a) [15 puntos] Descargar la página y extraer sus tablas HTML utilizando `requests` y `pandas` (o `BeautifulSoup`, si fuera necesario).  

    **Pista:** Use `pd.read_html()` para extraer tablas HTML. Busque tablas con la clase CSS wikitable.

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Cantones_de_Costa_Rica'
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})

webpage = urlopen(req).read()
# convertir bytes -> texto usando UTF-8
html = webpage.decode("utf-8")

pagina = BeautifulSoup(webpage, 'html.parser')
print(pagina.title.string)
```

## Parte 2
- b) [20 puntos] Identificar y extraer la tabla principal que contiene la información de los cantones. La tabla final debe incluir, como mínimo, columnas equivalentes a provincia, cantón, ciudad cabecera, año de creación, área y población.

    **Pista:** No todas las tablas son útiles. Implemente una función que verifique si
una tabla contiene las columnas esperadas (busque palabras clave como cantón,
provincia).

```python
# table class=wikitable sortable jquery-tablesorter
# Provincia, Canton, Ciudad Cabecera, año de creación, área y población
etqTable = pagina.find("table", {'class':'wikitable sortable'})
# print(etqTable)
print(etqTable.text.strip()[:500])
```

```python
tablas = pd.read_html(StringIO(html))
tablas
```

```python
def es_tabla_cantones(df):
    
    columnas = " ".join([str(c) for c in df.columns]).lower()
    
    if "provincia" in columnas and "cantón" in columnas:
        return True
    
    return False


tabla_cantones = None

for tabla in tablas:
    if es_tabla_cantones(tabla):
        tabla_cantones = tabla
        break

tabla_cantones.head()
```

```python
df = tabla_cantones.copy()
df.shape
```

```python
df.columns.tolist()
```

```python
df.columns = df.columns.get_level_values(1) # Aplanamos las columnas, eliminacion de multi-index
df.head(3)
```

```python
# Eliminacion de columnas que no necesitamos
df = df.drop(columns=['Cantón', 'Altitud (m.s.n.m)','IDH 2020', 'Partido', 'Nombre', 'Est. Dens. 2022','Censo 2011'])
df.sample(5)
```

## Parte 3
- c) [10 puntos] Realizar limpieza básica de los datos: eliminar filas completamente vacías, normalizar nombres de columnas y dejar la tabla lista para análisis en pandas.

```python
print(repr(df['Est. 2022'].iloc[0]))
```

```python
# limpiar espacios
df.columns = df.columns.str.strip()
df.head(3)
```

```python
df = df.dropna(how="all")
df.info()
```

```python
# Eliminacion de caracter espcial \xa0 entre numeros,
# se pasa a sting se eliminan espacios y luego se pasa a int
df['Est. 2022'] = (df['Est. 2022']
                   .astype(str)
                   .str.replace(r"\s+", "", regex=True)
                   .astype(int)
)
df
```

```python
df.info()
```

```python
# Renombrar columnas
df = df.rename(columns={'Cantón.1':'Cantón', 'Est. 2022':'Población'})
```

```python
# Filas sin datos faltantes
df.isna().sum()
```

## Parte 4
- d) [5 puntos] Guardar el resultado final en un archivo CSV llamado `cantones_cr.csv`.

```python
df.head(4)
```

```python
df.to_csv("cantones_cr.csv", index=False, encoding="utf-8")
```
